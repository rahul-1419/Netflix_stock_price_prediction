```python
for i in df:
    plt.figure(figsize=(6,4))
    sns.scatterplot(data=df, x=i, y='Close')
    plt.ylabel('Close Price')
    plt.xlabel(f'{i}')
    plt.title(f'{i} vs Close Price')
    plt.show()
```

```python
for i in df:
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=i, kde=True, bins=30)
    plt.xlabel(f'{i}')
    plt.title(f'Histogram of {i}')
    plt.show()
```
